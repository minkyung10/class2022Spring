### Constructing a DataFrame using dict

CSV 파일을 통해서가 아니라, 바로 dataframe을 만드는 방법

```python
# Some input data - a small sample of the iris dataset
data = {'sepal_length': [6.9, 6.9, 4.8, 5.4, 4.6],              #key와 value가 나열되어 있음
        'sepal_width': [3.2, 3.1, 3.4, 3.0, 3.6],               #value에는 list의 형태로 들어와야 함
        'petal_length': [5.7, 5.1, 1.9, 4.5, 1.0],
        'petal_width': [2.3, 2.3, 0.2, 1.5, 0.2],
        'species': ['virginica', 'virginica', 'setosa', 'versicolor', 'setosa']}

df = pd.DataFrame(data)             #dict를 입력으로 넣어주면 dataframe을 만들어준다
df                                  #엑셀 형태로 생긴 이 변수를 dataframe이라고 함
```

### Constructing a DataFrame using pd.Series

```python
sepal_length = pd.Series([6.9, 6.9, 4.8, 5.4, 4.6])
sepal_width = pd.Series([3.2, 3.1, 3.4, 3.0, 3.6])
petal_length = pd.Series([5.7, 5.1, 1.9, 4.5, 1.0])
petal_width = pd.Series([2.3, 2.3, 0.2, 1.5, 0.2])
species = pd.Series(['virginica', 'virginica', 'setosa', 'versicolor', 'setosa'])

#dict의 형태를 사용
df = pd.DataFrame({'sepal_length': sepal_length, 'sepal_width': sepal_width, 'petal_length': petal_length, 'petal_width': petal_width, 'SPECIES': species})
df
```

dataframe 만드는 방법 3가지

1. CSV파일에서 불러와서 변환

2. dict파일을 통으로 만들어서 만드는 방법

3. series를 써서 column의 내용 부분만 먼저 만든 다음에 그 다음에 이름 연결

**-만들어진 dataframe에서 일부 정보에만 접근하는 방법-**

### Changing column names

```python
df.columns
```

species 부분만 대문자인거 다시 소문자로 바꾸고 싶을 때

```python
# Change the columns
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
df
```

### Accessing columns

원하는 정보만 가져오고 싶을 때

```python
# Access one column 
df.loc[:, 'sepal_length']        # 컴마의 왼쪽은 행, 오른쪽은 열
```

```python
# Access one column (short-hand)
df['sepal_length']
```

```python
# Access one column (shortest-hand)
df.sepal_length
```

```python
# Access two columns
df.loc[:, ['sepal_length', 'sepal_width']]
```

```python
# Access two columns (short-hand)
df[['sepal_length', 'sepal_width']]
```

### Accessing rows

```python
# Access one row 
df.loc[3, :]
```

```python
# Access two rows
df.loc[[3, 0], :]     #3번째 행과 0번째 행을 같이 가져오는 것
```

```python
# Access a range of rows
df.loc[2:5, :]
```

```python
# Filter keeping only rows with `sepal_length` > 5
df[df.sepal_length > 5]        #df에서 sepal_length가 5보다 큰 경우만 불러와라
```

### Accessing a cell

```python
df.loc[:, 'sepal_length'][4]    #한 column 통으로 가져온 다음에 그 중에 4번째 cell에만 access
```

```python
df['sepal_length'][4]
```

```python
df.sepal_length[4]
```

### Adding/Removing new columns

```python
df['new_column'] = 1
df
```

```python
df['new_column'] = [1, 2, 3, 4, 5]
df
```

```python
df = df.drop('new_column', axis=1) # axis=1: column(열)  #axis=0: row(행)
df
```

### Adding/Removing new rows

```python
df.loc[5, :] = ''    #안에 아무것도 없는 string
df
```

```python
df = df.drop(5, axis=0) # axis=0: row
df
```
